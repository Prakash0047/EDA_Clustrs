# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
data_F= sc.fit_transform(REL)
```

```python
data_F.mean()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,20):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_F)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,20), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=7,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data_F) 
```

```python
#visualizing the clusters
plt.scatter(data_F[y_kmeans==0,0], data_F[y_kmeans==0,1],s = 50,c = 'red',label='cluster 1')
plt.scatter(data_F[y_kmeans==1,0], data_F[y_kmeans==1,1],s=50,c='blue',label='Cluster 2')
plt.scatter(data_F[y_kmeans==2,0], data_F[y_kmeans==2,1], s=50,c='green',label='Cluster 3')
plt.scatter(data_F[y_kmeans==3,0], data_F[y_kmeans==3,1], s=50,c='magenta',label='Cluster 4')
plt.scatter(data_F[y_kmeans==4,0], data_F[y_kmeans==4,1], s=50,c='black',label='Cluster 5')
plt.scatter(data_F[y_kmeans==5,0], data_F[y_kmeans==5,1], s=50,c='brown',label='Cluster 6')
plt.scatter(data_F[y_kmeans==6,0], data_F[y_kmeans==6,1], s=50,c='purple',label='Cluster 7')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=80,c='yellow',label='Centroids',alpha=1.0)
plt.xlabel
plt.legend()
plt.show()
```
